```python
import pandas as pd
import numpy as np
```

```python
content=pd.read_csv("Content.csv")
content.head()
```

```python
content.shape
```

```python
content=content.drop(['Unnamed: 0','User ID','URL'],axis=1)
content.head()
```

```python
content.dtypes
```

```python
content[['Type','Category']]=content[['Type','Category']].astype(str)
content.dtypes
```

```python
reactions=pd.read_csv("Reactions.csv")
reactions.head()
```

```python
reactions.shape
```

```python
reactions=reactions.drop(['Unnamed: 0','User ID'],axis=1)
```

```python
reactions.isnull()
```

```python
reactions=reactions.dropna()
reactions.shape
```

```python
reactions.head()
```

```python
reactions.dtypes
```

```python
reactions['Type']=reactions['Type'].astype(str)
```

```python
reactiontypes=pd.read_csv("ReactionTypes.csv")
reactiontypes.head()
```

```python
reactiontypes.shape
```

```python
reactiontypes.isnull()
```

```python
reactiontypes.dtypes
```

```python
reactiontypes[['Type','Sentiment']]=reactiontypes[['Type','Sentiment']].astype(str)
reactiontypes.dtypes
```

```python
reactiontypes=reactiontypes.drop(['Unnamed: 0','Sentiment'],axis=1)
```

```python
reactiontypes.head()
```

```python
content.to_csv('CleanContent.csv')
reactions.to_csv('CleanReactions.csv')
reactiontypes.to_csv('CleanReactiontypes.csv')
```

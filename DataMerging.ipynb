```python
import pandas as pd
from google.colab import files
```

```python
content=pd.read_csv("CleanContent.csv",index_col=0)
reactions=pd.read_csv("CleanReactions.csv",index_col=0)
reactiontypes=pd.read_csv("CleanReactiontypes.csv",index_col=0)
```

```python
reactions.head()
```

```python
content.head()
```

```python
reactiontypes.head()
```

```python
reactions1=pd.merge(reactions,reactiontypes,on='Type')
```

```python
reactions1.head()
```

```python
print(reactions.shape)
print(reactions1.shape)
```

```python
content.shape
```

```python
reactions2=pd.merge(reactions1,content,on='Content ID')
```

```python
print(reactions.shape)
print(reactions1.shape)
print(reactions2.shape)
```

```python
reactions2.head()
```

```python
reactions2=reactions2.rename({'Type_x':'Reaction_type','Type_y':'Content_type'},axis=1)
```

```python
reactions2.head()
```

```python
df=reactions2.groupby(['Category']).agg({'Score':'sum'}).sort_values()
```

```python
df.head()
```

```python
df=df.sort_values(by='Score',ascending=False)
```

```python
df
```

```python
df1=df[0:5]
```

```python
df1.to_csv("top_5.csv")
```

```python
reactions2.to_csv("cleaneddataset.csv")
```
